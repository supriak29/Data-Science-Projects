Before we can do anything about plotting histograms, we need to make sure that the columns that we offer to the histogram is actually plottable, meaning that we don't care about categorical variables. So those have to be removed and any variables with Na have to be removed

```python
dataset2 = dataset.drop(columns = ['Entry_id', 'pay_schedule', 'e_signed'])
```

```python
dataset2.columns
```

##### Plotting

```python
fig = plt.figure(figsize=(15,12)) # setting the figure size;

plt.suptitle('Histograms of Numerical Columns', fontsize=20) # title of cols. So of the entire image, which is histograms on the vertical column

for i in range(dataset2.shape[1]):           # For each of the columns,
    plt.subplot(6, 3, i+1)                   # we're going to create a new subplot
    f = plt.gca()                            # We're going to plot it: lets the current axes so that you can draw on it directly
    f.set_title(dataset2.columns.values[i])  # we're going to set the title of that plot to the particular column that plot is representing
    
    vals = np.size(dataset2.iloc[:, i].unique())  # bias variable
    # safe guard to make sure that our plots are not taking too long to to display
    if vals >= 100:
        vals = 100
    
   
    plt.hist(dataset2.iloc[:, i], bins=vals, color='#3F5D7D')  # plot histogram
plt.tight_layout(rect=[0, 0.03, 1, 0.95])  # set up a layout rectangle so that nothing overlaps

```

##### Bias Variable - vals

<html>
<p>
In the previous videos in which we did Histograms, we created the values variable. 
The bias variable pretty much is a is all of the unique values here in your column that you're trying to plot
So if you have a hundred years in your your age column, then the histogram is going to support up to 100 little ticks.
<br><br>
Now, if you have a lot of a lot of different numbers, a big range, especially for continuous variables like the risk scores.
So those can take a very unique value for all of the users.
So in those cases, you're going to end up having like thousands and thousands of different possible values here in 
balls and the PLT, the histogram function is going to struggle to pretty much create a bin for all of those possible values.
<br><br>
And although it is going to work, it's going to take a long time to plot and it may even crash your system 
depending on how powerful your system is.
<br><br>
So that is why I created this extra variable here called the bias variable.
Well, it's already the bias variable, but this is a limit on the bias variable.
<br><br>
So what it does is that if there is more than 100 unique values, for example, in the risk score column,
then the number of bins are going to default to 100.
That way we make sure that no histogram that is plotted in this big image is going to have more than
100 bins per histogram, and that it's not going to struggle to plot all of these values.
</p>

## Correlation Plot

We're going to quickly plot the correlation plot between all the features that we have explored and the response variable.

```python
# Correlation with Response Variable (Note: Models like RF are not linear like these)

# dataset2.corrwith: This is going to create a set of correlations between each of the columns and its argument
# (dataset.e_signed): Its argument, of course, is the response variable, which is what we want to analyze

dataset2.corrwith(dataset.e_signed).plot.bar(
    figsize = (20,10),
    title = 'Correlation with E-signed',
    fontsize = 15,
    rot = 45,
    grid = True
)

# .plot.bar(: plot these results
# figure size = depends on your screen
# rotation argument, which is just so that I can read the columns better; ticks of the x axis 
# grid argument, which gives it a grid on the background
```

## Correlation Matrix

We're going to jump into the correlation matrix, see the relationship between the fields with each other, and try to understand if we need some kind of feature engineering or not. <br>
We also can use our previous plots that we have used to gauge whether or not we need some feature engineering or we need to remove some columns.

```python
sn.set(style='white')
```

```python
# compute the correlation matrix
corr = dataset2.corr()

# this correlation method gives us the correlation of each field with each other, 
# so it is a lot more robust than the core weight function that we used previously
```

```python
# generate a mask for upper triangle
mask = np.zeros_like(corr, dtype=np.bool_)
mask[np.triu_indices_from(mask)] = True
```

```python
# setup the matplotlib figure
f, ax = plt.subplots(figsize=(18, 15))

# generate a custom diverging colormap
cmap = sn.diverging_palette(220, 10, as_cmap=True)

# Draw heatmap with mask and aspect ratio
sn.heatmap(corr, mask=mask, cmap=cmap, vmax=3, center=0, square=True, linewidths=.5, cbar_kws={"shrink": .5})
```
